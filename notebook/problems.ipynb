## Binomial distribution

In probability theory and statistics, the binomial distribution is the discrete probability distribution that gives only two possible results in an experiment; for example, when you flip a coin, it can be either head or tails.

```python
from scipy.stats import binom

# Define all the probability functions related to this distribution

def dbinom(x, mean, prob = 0.5):
    """
    Calculates the point estimate of the binomial distribution
    """
    result = binom.pmf(k = x, n = mean, p = prob, loc = 0)

    return result

def pbinom(q, size, prob = 0.5):
    """
    Calculates the cumulative of the binomial distribution
    """
    result = binom.cdf(k = q, n = size, p = prob, loc = 0)

    return result

def qbinom(p, size, prob = 0.5):
    """
    Calculates the quantile function from the binomial distribution
    """
    result = binom.ppf(q = p, n = size, p = prob, loc = 0)

    return result

def rbinom(n, size, prob = 0.5):
    """
    Generates random variables from the binomial distribution
    """
    result = binom.rvs(n = size, p = prob, size = n)

    return result
```

```python
# Binomial(10, 0.2) distribution, where n = 10 is the number of trials and p = 0.2 is the probability of success in a single trial
import numpy as np

np.random.seed(42)

print(f"Probability that a Binomial(10, 0.2) takes the value 2: {dbinom(2, size = 10, prob = 0.2)}")
print(f"Probability that a Binomial(10, 0.2) will take a value less than 2: {pbinom(2, size = 10, prob = 0.2)}")
print(f"Which value of a Binomial(10, 0.2) has a cumulative probability of 0.9? {qbinom(0.9, size = 10, prob = 0.2)}")
print(f"Generate 2000 random values from a Binomial(10, 0.2) distribution: {rbinom(2000, size = 10, prob = 0.2)}")
```

```python
import matplotlib.pyplot as plt

np.random.seed(42)

size = 10
prob = 0.2
data = rbinom(2000, size = 10, prob = 0.2)

pmf = dbinom(data, size = size, prob = prob)
cdf = pbinom(data, size = size, prob = prob)
ppf = qbinom(data, size = size, prob = prob)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pmf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

Normal Distribution

```python
from scipy.stats import norm

def dnorm(x, mean = 0, std = 1):
    return norm.pdf(x, loc = mean, scale = std)

def cnorm(x, mean = 0, std = 1):
    return norm.cdf(x, loc = mean, scale = std)

def qnorm(x, mean = 0, std = 1):
    return norm.ppf(x, loc = mean, scale = std)

def rnorm(size, mean = 0, std = 1):
    return norm.rvs(size = size, loc = mean, scale = std)
```

```python
import numpy as np

np.random.seed(42)

print(f"Probability that a Normal(0, 1) takes the value 2: {dnorm(2, mean = 0, std = 1)}")
print(f"Probability that a Normal(0, 1) will take a value less than 2: {cnorm(2, mean = 0, std = 1)}")
print(f"Which value of a Normal(0, 1) has a cumulative probability of 0.9? {qnorm(0.9, mean = 0, std = 1)}")
print(f"Generate 2000 random values from a Normal(0, 1) distribution: {rnorm(2000, mean = 0, std = 1)}")
```

```python
import matplotlib.pyplot as plt

np.random.seed(42)

mean = 0
std = 1
data = (np.arange(-10,10,0.01))

pmf = dnorm(data, mean = mean, std = std)
cdf = cnorm(data, mean = mean, std = std)
ppf = qnorm(data, mean = mean, std = std)
rvs = rnorm(2000, mean = mean, std = std)

fig, axis = plt.subplots(4, 1, figsize = (5, 7))

axis[0].plot(data, pmf, "r-")
axis[1].plot(data, cdf, "b-")
axis[2].plot(data, ppf, "g-")
axis[3].hist(rvs, bins = 500, color='yellow')

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

Uniform Distribution

```python
from scipy.stats import uniform

def dunif(x, a = 0, b = 1):
    return uniform.pdf(x, loc = a, scale = b)

def cunif(x, a = 0, b = 1):
    return uniform.cdf(x, loc = a, scale = b)

def qunif(x, a = 0, b = 1):
    return uniform.ppf(x, loc = a, scale = b)

def runif(size, a = 0, b = 1):
    return uniform.rvs(size = size, loc = a, scale = b)
```

```python
import numpy as np

np.random.seed(42)

print(f"Probability that a Uniform[0, 1] takes the value 0.3: {dunif(0.3, a = 0, b = 1)}")
print(f"Probability that a Uniform[0, 1] will take a value less than 0.3: {cunif(0.3, a = 0, b = 1)}")
print(f"Which value of a Uniform[0, 1] has a cumulative probability of 0.9? {qunif(0.9, a = 0, b = 1)}")
print(f"Generate 2000 random values from a Uniform[0, 1] distribution: {runif(2000, a = 0, b = 1)}")
```

```python
import matplotlib.pyplot as plt

np.random.seed(42)

a = 0
b = 1
data = runif(2000, a = a, b = b)

pmf = dunif(data, a = a, b = b)
cdf = cunif(data, a = a, b = b)
ppf = qunif(data, a = a, b = b)

fig, axis = plt.subplots(4, 1, figsize = (5, 7))

axis[0].plot(data, pmf, "r-")
axis[1].plot(data, cdf, "b-")
axis[2].plot(data, ppf, "g-")
axis[3].hist(data, bins = 200, color='yellow')

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```
